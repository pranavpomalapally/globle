```python
def globle_shade_countries(geolocator, locs, dists, country, answer):
    while (True):
        print("If you are dumb, please type: I am Ajay Balaraman and I give up.")
        user_guess = input("(Player 1) Guess a country: ")

        if user_guess == 'quit':
            print("you loser")
            break

        if user_guess == 'I am Ajay Balaraman and I give up.':
            print("you loser")
            print("It's ", country)
            break

        location = geolocator.geocode(user_guess, language="en")
        if not location:
            print("that was such a bad guess I'm gonna skip that")
            continue

        guess = (location.latitude, location.longitude)
        dist = geodesic(answer, guess).km

        locs.append(location.address)
        dists.append(dist)
        d = {'Country': locs, 'Distance Away': dists}
        df = pd.DataFrame(d)

        fig = px.choropleth(data_frame=df,
                        locations='Country', 
                        locationmode="country names",
                        range_color=(0, 15000),
                        color_continuous_scale='RdBu',
                        color = 'Distance Away')
        fig.show()
        print(df)
        print('\n')

        if dist == 0:
            print("you are a goat.")
            break
```

```python
def globle_point_flags(geolocator, locs, dists, country, answer):
    map = folium.Map(location=[0, 0], zoom_start = 2,
                                zoom_control=True,
                                scrollWheelZoom=False,
                                dragging=True)
    while (True):
        print("If you are dumb, please type: I am Ajay Balaraman and I give up.")
        user_guess = input("(Player 1) Guess a country: ")

        if user_guess == 'quit':
            print("you loser")
            break

        if user_guess == 'I am Ajay Balaraman and I give up.':
            print("you loser")
            print("It's ", country)
            break

        location = geolocator.geocode(user_guess)
        if not location:
            print("that was such a bad guess I'm gonna skip that")
            continue

        guess = (location.latitude, location.longitude)
        dist = geodesic(answer, guess).km
        
        locs.append(location.address)
        dists.append(dist)
        d = {'Country': locs, 'Distance Away': dists}
        df = pd.DataFrame(d)

        map.add_child(folium.Marker(location=[location.latitude, 
                                    location.longitude], 
                                    popup='Default popup Marker1',
                                    tooltip=dist))
        display(map)
        print(df)
        print('\n')
        
        if dist == 0:
            print("you are a goat.")
            break
```
